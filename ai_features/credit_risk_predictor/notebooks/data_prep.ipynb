```python
"""
Select only relevant columns for credit risk modeling.
Keeps financial and repayment-related features required for
feature engineering and removes all unnecessary attributes.
"""

relevant_columns = [
    "LIMIT_BAL",
    "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6",
    "BILL_AMT1", "BILL_AMT2", "BILL_AMT3",
    "BILL_AMT4", "BILL_AMT5", "BILL_AMT6",
    "PAY_AMT1", "PAY_AMT2", "PAY_AMT3",
    "PAY_AMT4", "PAY_AMT5", "PAY_AMT6",
    "default.payment.next.month"
]

credit_df = credit_df[relevant_columns]
```

```python
credit_df
```

```python
"""
Rename key columns for clarity and consistency
"""

credit_df = credit_df.rename(columns={
    "LIMIT_BAL": "credit_limit",
    "default.payment.next.month": "default"
})
```

```python
# Feature Engineering
```

```python
"""
Create core financial features from bill and payment data
"""

# Current outstanding balance (most recent bill)
credit_df["current_outstanding_balance"] = credit_df["BILL_AMT1"]

# Average bill amount
credit_df["avg_bill_amount"] = credit_df[
    ["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]
].mean(axis=1)

# Total paid amount
credit_df["total_paid"] = credit_df[
    ["PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]
].sum(axis=1)
```

```python
credit_df
```

```python
"""
Create repayment strength indicator
"""

total_bill = credit_df[
    ["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]
].sum(axis=1)

credit_df["payment_ratio"] = credit_df["total_paid"] / (total_bill + 1)  # avoid division by zero
```

```python
"""
Create delay-related behavioral features
"""

pay_cols = ["PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]

# Number of late payments
credit_df["num_late_payments"] = (credit_df[pay_cols] > 0).sum(axis=1)

# Average delay
credit_df["avg_delay"] = credit_df[pay_cols].mean(axis=1)

# Maximum delay
credit_df["max_delay"] = credit_df[pay_cols].max(axis=1)

# Most recent delay
credit_df["recent_delay"] = credit_df["PAY_0"]
```

```python
credit_df
```

```python
"""
Create credit utilization feature
"""

credit_df["credit_utilization"] = (
    credit_df["current_outstanding_balance"] / (credit_df["credit_limit"] + 1)
)
```

```python
credit_df
```

```python
"""
Remove raw columns after feature engineering
"""

raw_columns = [
    "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6",
    "BILL_AMT1", "BILL_AMT2", "BILL_AMT3",
    "BILL_AMT4", "BILL_AMT5", "BILL_AMT6",
    "PAY_AMT1", "PAY_AMT2", "PAY_AMT3",
    "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"
]

credit_df = credit_df.drop(columns=raw_columns)
```
